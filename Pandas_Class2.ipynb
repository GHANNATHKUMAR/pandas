## 1 Inspecting Data
Quick ways to understand your dataset:
- `head(n)` → first *n* rows (default 5)
- `tail(n)` → last *n* rows
- `info()` → columns, non-null counts, and dtypes
- `shape` → (rows, columns)
- `dtypes` → data type of each column

```python
import pandas as pd

# Create a small DataFrame
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)
b.head(2)
```

```python
b.head() #by defdault give 5 row
```

```python
b.head(3) #first 3 rows
```

```python
b.tail(3) #last three rows
```

```python
b.info() #non null means , value is not null
```

```python
b.shape #5 rows and 3 columns
```

```python
print(b) #here jaisa value hai vaisa hi hai data types , in numpy agar ek chiz dusra rahta tha toh sari ki sari chiz object ho jati thi like if a element is tring whole array ka data type object ho jata tha
```

```python
b.dtypes #harek columns ka datatypes
```

## 2 Selecting Columns & Rows
Ways to access parts of a DataFrame:
- Single column → `df['col']`
- Multiple columns → `df[['col1','col2']]`
- Label-based selection → `df.loc[row_label, col_label]`
- Integer-index selection → `df.iloc[row_idx, col_idx]`
Tip: `:` means "all rows/columns" in that dimension.

```python
print(b)
b['Name'] ##now here we will select here jo chiz pahle select karenge vo column honge becuase dabdaba jayda rahta hai in some cases  like here
```

```python
b[['Name','Age']] #fancy indexing
```

```python
b.loc[0] #row number 0 ka aaya and series ban gaya
```

```python
b.loc[0,"Score"] #0 index pe Score ka value aa gaya hai  which is 88
```

```python
print(b)
b.loc[1:3,["Name","Age"]]  #1 se 3 including , slicing , columns name #note that 1:3 means 1 se 3 location le raha hai ye but iloc thoda sa different kaam karta hai
```

```python
b.iloc[0] #same thing aaya but iloc is used when rows like other hai jaise 1,2,3,4,5 jaise ho toh iloc index locate karta hai like 0 for 1 1 for 2 like that but yahan 0 se hi chalu hai toh no problem
```

## 4 Renaming Columns & Setting an Index
- Rename columns → `df.rename(columns={'old':'new'}, inplace=False)`
- Set a column as index → `df.set_index('column_name', inplace=False)`
- Reset index → `df.reset_index(inplace=False)`
 If you don't use `inplace=True`, remember to assign the result back to a variable.

```python

# c=b.rename(columns={"Score":"Marks"}) #we have not changed our original dataset that's why we kept in c
# print(c)
```

```python
#inplace true means copy pe na lagke pure dat ape hi apply ho jaye
b.rename(columns={"Score":"Marks"}) #here copy only without inplace=True , and it will apply only on copy , our main original will not change
b #but still score so inplace when true then it will apply on original data
```

```python
b.rename(columns={"Score":"Marks"},inplace=True) #ye karne par hamara original dataset change ho jayenge from score to marks , # it will apply on main data
b
```

```python
b.rename(columns={"Marks":"Score"},inplace=True) #again changed original
b
```

c=b.set_index("Name") #and Name as index points
c # and name came as a index
c.loc["Alice"] #and we got alice data as a series  

```python
b.set_index("Name",inplace=True) #so mera b ko bigad diya , matlab original dataset ko change kar diya , aur agar hamein ise dubara original mein lana hai toh we will use reset
b
```

```python
b.reset_index(inplace=True)
b
```
