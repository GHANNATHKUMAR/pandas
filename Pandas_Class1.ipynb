## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
a=[10,20,30,40,50]
s=pd.Series(a) # so ek column ban gaya , i.e ek series ban gaya yani can say only one row
s
````

````python
a=[10,20,30,40,50]
s=pd.Series(a,name="marks")
s
````

````python
d={"Akarsh":90,"Ghannath":100,"Rohan":60,"Shivam":70,"Ayush":80}
s=pd.Series(d,name="Marks")
s
````

### Creating DataFrame from NumPy Arrays

````python
a=np.array([[1,2,3],
           [3,4,5],
           [6,7,8]])
df=pd.DataFrame(a)
df
#BUT HAM LOGON KO AISE DATA BANANE KI JARURAT NAHI HOTI , HAMLOGON KE PASS CSV FILES , EXCEL FILES HOTE HAIN AND MANY RESOURCES JAHAN SE HAM DOWNLOAD KAR SAKTE HAIN LIKE KAGGLE ETC
#Kaggle is a popular Google-owned online platform and community for data science and machine learning enthusiasts. It hosts competitions to solve real-world problems, provides a vast repository of datasets, offers a browser-based, no-setup environment for writing code (Notebooks), and serves as a hub for learning and collaboration
````
