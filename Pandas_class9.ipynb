## 1. Applying Custom Functions with `apply()` / `map()`
- `apply()` lets you apply a custom function across rows or columns.
- `map()` is for applying functions element-wise to a Series.
- Both are handy for transformations, calculations, and cleaning data.

```python
import numpy as np
import pandas as pd
data={
  "Name":["Alice","Bob","Charlie","David","Emma"],
  "Math":[88,75,92,60,85],
  "Science":[90,80,78,70,95]
}
df=pd.DataFrame(data)
df
```

```python
df['ToatalMarks']=df[['Math','Science']].apply(lambda row:row.sum(),axis=1)
```

```python
df
```

```python
col_meandf=df[['Math','Science']].apply(lambda  col:col.mean()) #here by default axis=0
col_meandf
```

```python
def to_grade(s):#custom function
   if s>=90:
    return "A"
   elif s>=80:
    return "B"
   elif s>=70:
    return "C"
   else:
    return "D"
df['MathsGrade']=df['Math'].apply(to_grade)
df
```

```python
df['Name']=df['Name'].map(str.lower)
df
```

```python
df['Math']=df['Math'].map(lambda x:x+5)
df
```
