### Create a new column score_level
"High" if total_score ≥ 90 , "Medium" if 70 ≤ total_score < 90, "Low" if total_score < 70

```python
def level(x):
  if x>=90:
    return "High"
  elif x<90 and x>=70:
    return "Medium"
  else :
    return "low"
```

```python
df['scoring_level']=df['total_score'].apply(level)
```

### Create a new column study_category
"Light" if weekly_self_study_hours < 10,
"Moderate" if 10 ≤ hours ≤ 20,
"Heavy" if hours > 20

```python
def category(x):
  if x>20:
    return "Heavy"
  elif x>10:
    return "Moderate"
  else:
    return "Light"
```

```python
df['student_category']=df['weekly_self_study_hours'].apply(category)
```
