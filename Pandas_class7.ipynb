## 1. Using `concat()` for Stacking Data
- `pd.concat()` combines DataFrames vertically (rows) or horizontally (columns).
- Use `axis=0` for vertical stacking (default), `axis=1` for horizontal.
- `ignore_index=True` can reset the index after concatenation.

```python
import pandas as pd

# Dataset 1: Product details
df1=pd.DataFrame({
    "ProductID":[101,102,103],
    "ProductName":["Pen","Notebook","Marker"],
    "Price":[10,50,30]
})

# Dataset 2: More products (same columns as df1 → good for vertical stacking)
df2=pd.DataFrame({
     "ProductID":[104,105],
     "ProductName":["Pencil","Eraser"],
     "Price":[5,8]
 })
# Dataset 3: Additional info (different columns → good for horizontal stacking)
df3=pd.DataFrame({
     "Stock":[100,60,80],
     "Rating":[4.5,4.7,4.8]
 })
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat((df1,df2),axis=0)
```

```python
pd.concat([df1,df3],axis=0) #row increase
```

```python
pd.concat([df1,df3],axis=1) #columns increase
```

## 2. Using `merge()` for SQL-Style Joins
- `merge()` is similar to SQL joins.
- Specify `on` or `left_on`/`right_on` to choose key columns.
- Join types: `inner` (default), `left`, `right`, `outer`.

```python
products = pd.concat([df1, df2], ignore_index=True)
# ignore_index=True is recommended to avoid duplicate index 0, 1
```

```python
sales = pd.DataFrame({
    "ProductID": [101, 102, 104, 106],
    "UnitsSold": [500, 200, 150, 50]
})
```

```python
pd.merge(products,sales)
```

```python
pd.merge(products,sales,on="ProductID",how="inner") #how inner join and on ProductID  #both common
```

```python
pd.merge(products,sales,on="ProductID",how="left")  #only focused on products i.e left
```

```python
pd.merge(products,sales,on="ProductID",how="right") #focused on right
```

```python
pd.merge(products,sales,on="ProductID",how="outer" )   #all will come  both 105 and 106 will also come
```

## 3. Using `join()` with Index Alignment
- `join()` joins DataFrames on their index or on a key column.
- By default, joins on index; can specify `on='col'`.
- Useful for adding columns to an existing DataFrame.

```python
products = pd.concat([df1, df2], ignore_index=True)
products
```

```python
stock=pd.DataFrame({
    "Stock":[100,60,80,120,150 ]
},index=[101,102,103,104,105])
stock
```

```python
p_dataset=products.set_index('ProductID')
p_dataset
```

```python
p_dataset.join(stock) # when same index come in picture
```

## 4. Practical Examples
- Combine sales and revenue data to see performance.
- Merge students with marks to link records.
- Practice with realistic datasets to understand alignment and missing data handling.

```python
import pandas as pd

# Sales data
sales = pd.DataFrame({
    "ProductID": [201, 202, 203, 204],
    "UnitsSold": [500, 300, 400, 250]
})

# Revenue data (missing ProductID 204, extra 205)
revenue = pd.DataFrame({
    "ProductID": [201, 202, 203, 205],
    "Revenue": [7500, 4500, 6800, 1200]
})
print("Sales Data:")
print(sales)
print("\nRevenue Data:")
print(revenue)
```

```python
merged_df=pd.merge(sales,revenue ,on="ProductID",how="outer") #some null value will also come becuase there are  join on outer
merged_df
```

```python
merged_df=pd.merge(sales,revenue ,on="ProductID",how="inner")
merged_df
```

```python
merged_df=pd.merge(sales,revenue ,on="ProductID",how="left")
merged_df
```

```python
merged_df=pd.merge(sales,revenue ,on="ProductID",how="right")
merged_df
```
