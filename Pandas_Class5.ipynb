## 1. Sorting Data
- Use `sort_values(by='col')` to sort by column values.
- Use `ascending=False` for descending order.
- Sort by multiple columns by passing a list.
- `sort_index()` sorts by row or column index.

```python
import pandas as pd
data = {
    "Product": ["Pen", "Notebook", "Pencil", "Marker", "Eraser", "Sharpener"],
    "Price": [10, 50, 5, 30, 8, 12],
    "Stock": [100, 60, 200, 80, 150, 120],
    "Rating": [4.5, 4.7, 4.2, 4.8, 4.3, 4.4]
}
df=pd.DataFrame(data)
df
```

```python
df.sort_values(by='Price')
```

```python
df.sort_values(by='Price',ascending=False)
```

```python
# df.sort_values(by='Price',ascending=False,inplace=True) #original table will update
```

```python
#sort by multiple columns (price ascending and then stock descending ) #note that it will happen when we will have data such type
df.sort_values(by=['Price','Stock'],ascending=[True,False])
```

## 2. Descriptive Statistics
- `mean()`, `median()`, `mode()` for central tendency.
- `std()` for standard deviation.
- `describe()` for a summary of stats (count, mean, std, min, quartiles, max).

```python
df.mean(numeric_only=True) #sabka column wise mean de dega
```

```python
df.median(numeric_only=True)
```

```python
df.mode() #mode isiliye nahi aa raha hai kyunki sari values ek ek bar hi hai
```

```python
df.std(numeric_only=True)
```

```python
df.describe() #all values with 5 number summary
```

## 3. Counting Values
- `value_counts()` shows the frequency of unique values in a Series.
- Use `normalize=True` to see percentages instead of counts.

```python
df
```

```python
df['Product'].value_counts() #value unique values
```

```python
df['Product'].value_counts(normalize=True) #all in percentage in this case in all 1/6(float)  *100
```

## 4. Correlation & Covariance
- `corr()` computes correlation between numeric columns.
- `cov()` computes covariance.
- Correlation values range from -1 (negative) to +1 (positive).

```python
#corelation and covariance
df.corr(numeric_only=True)
```

```python
df.cov(numeric_only=True)
```

```python
df['Price'].corr(df['Rating']) #82% corelation between price and rating
```
